## 5. Define the directories to save images

```python
Download_dir = upper_dir + '/' + '2_Source_raw_fits_image' + '/'
NVSS_dir = Download_dir + 'NVSS/'
# FIRST
FIRST_dir = Download_dir + 'FIRST/'
```

The function to download image of typical survey

```python
def image_download(RA, DEC, object_id, object_class, survey, pixel_scale, download_dir):
    #---------------------------------
    # Target: 
    # Download image with customized central RA/DEC/survey/image scale/image object id.
    #---------------------------------
    # Input:
    # 1. RA: host galaxy RA
    # 2. DEC: host galaxy DEC
    # 3. object_id: the recorded object id
    # 4. object_class: FR class of the object
    # 5. survey: the survey image waiting for download
    #
    # 6. download_dir: the upper dir to save the fits image
    # 7. pixel_scale: the widht/height of an downloaded image
    #
    # Return:
    # 
    #---------------------------------
    # Part 1. generate url
    coord_posi = str(RA) + ',' + str(DEC)
    print(coord_posi)
    source_url = SkyView.get_image_list(position=coord_posi,
                  survey=survey,scaling='Linear',pixels=pixel_scale)
    print(source_url)
    # Part 2. define image name and actually download it
    download_filename = download_dir + object_id + '_' + object_class +'.fits'
    urllib.request.urlretrieve(source_url[0], download_filename)
    print(f'The {survey} survey image of object {object_id} with RA {RA}, DEC {DEC} , have been downloaded via url {source_url} in name: {download_filename}.\n')
    return 
```
